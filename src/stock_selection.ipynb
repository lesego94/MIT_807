```python
import yfinance as yf
import pandas as pd

import getFamaFrenchFactors as gff
```

```python
# 1. Gather Stock Symbols
stocks = ["MSFT", "JNJ","NEM", "GOLD", "FCX", "BHP", "GSK", "AZN",
          "RIO", "ABBV", "MRK", "TSLA", "GOOGL", "WBA",  "MDT", "ISRG", "GILD", "SYK", "CI", "REGN", "VRTX", "ILMN",
          "AEM", "HL", "AG", "PAAS", "KGC", "FNV", "WPM", "SSRM", "BTG", "HMY","DHR" , "UNH","TMO","PFE", "LLY"]

start_date = '2009-01-01' 
end_date = '2015-12-25'


# Risk-free rate assumption (annualized). Adjust based on current economic conditions.
risk_free_rate = 0.02

```

```python
"MDT" in stocks
```

```python
#2. Fetch Historical Data
all_data = pd.DataFrame()
for stock in stocks:
    df = yf.download(stock, start=start_date, end=end_date).resample('M').last()['Adj Close']
    all_data[stock] = df

```

```python
monthly_returns = all_data.pct_change()
ff3_monthly = gff.famaFrench3Factor(frequency='m')
ff3_monthly.rename(columns={"date_ff_factors": 'Date'}, inplace=True)
ff3_monthly.set_index('Date', inplace=True)
rf = ff3_monthly['RF']
rf = pd.DataFrame(rf)
data = rf.join(monthly_returns,how='right')
data
```

```python
data.isnull().sum().sum()
```

```python
description = data.describe() 
new = description.T
new
```

# Sort stocks by Sharpe Ratio

```python
new["sharpe_ratio"] = ((new["mean"]-new["mean"]['RF'])/ new["std"])
sharpe_high=new.sort_values(by=['sharpe_ratio']).tail(15)
sharpe_high
```

```python
sharpe_filter=sharpe_high.index
sharpe_filter
```

```python
new_by_return=new.sort_values(by=['mean'])
new_by_return.loc[sharpe_filter].sort_values(by=['mean'])
```

```python
new_by_volatility=new.sort_values(by=['std'])
new_by_volatility=new_by_volatility.loc[sharpe_filter].sort_values(by=['std'])
new_by_volatility
```

```python
low_risk=new_by_volatility['std'].head(10).index
low_risk
```

```python
high_risk=new_by_volatility['std'].tail(10).index
high_risk
```
