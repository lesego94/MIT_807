```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
risk_parity = pd.read_csv('./Risk_parity_MTL_results.csv', index_col='Date', parse_dates=True)
risk_parity.index= pd.to_datetime(risk_parity.index)
risk_parity
```

```python
mvo_mtl = pd.read_csv('./MVO_MTL_results.csv', index_col='Date', parse_dates=True)
mvo_mtl.index= pd.to_datetime(mvo_mtl.index)
mvo_mtl
```

```python
benchmark_data= pd.read_csv('./Benchmark.csv', index_col='Date', parse_dates=True)
benchmark_data
```

## Include historical Results

```python
risk_parity_hist = pd.read_csv('./Historical/risk_parity_hist_results.csv', index_col='Date', parse_dates=True)
mvo_hist = pd.read_csv('./Historical/mvo_hist_results.csv', index_col='Date', parse_dates=True)
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(
    x=mvo_mtl.index,
    y=mvo_mtl['Portfolio_Value_rets'].cumsum(),
    mode='lines',
    name='MVO*',
    line=dict(width=1, color='blue', dash='solid')  # Solid blue line
))
fig.add_trace(go.Scatter(
    x=mvo_hist.index,
    y=mvo_hist['Portfolio_Value_rets'].cumsum(),
    mode='lines',
    name='MVO Historical',
    line=dict(width=1, color='blue', dash='dash')  # Solid blue line
))

fig.add_trace(go.Scatter(
    x=risk_parity.index,
    y=risk_parity['Portfolio_Value_rets'].cumsum(),
    mode='lines',
    name='Risk Parity*',
    line=dict(width=1, color='black', dash='solid')  # Solid black line
))
fig.add_trace(go.Scatter(
    x=risk_parity_hist.index,
    y=risk_parity_hist['Portfolio_Value_rets'].cumsum(),
    mode='lines',
    name='Risk Parity Historical',
    line=dict(width=1, color='black', dash='dash')  # Solid black line
))


fig.add_trace(go.Scatter(x=benchmark_data.index, y=benchmark_data['benchmark_rets'].cumsum(), name='Benchmark'))

#fig.update_layout(height=400, width=900, title='Strategy Overview', showlegend=True)

fig.update_layout(

    xaxis_title='Date',  # Label for the x-axis
    yaxis_title='Cumulative Returns',  # Label for the y-axis
    legend_title='',  # Title for the legend
    font=dict(
        family="Sans Serif",  # Font family
        size=18,  # Size of the text
        color="black"  # Color of the text
    ),
    height=500,  # Height of the figure in pixels
    width =900,  # Width of the figure in pixels
    showlegend=True,  # Show the legend (set to True)
    legend=dict(
        y=1, x=0,  # Position the legend at the top left
        xanchor='left', yanchor='top',
        bgcolor='rgba(0,0,0,0)'
  # Anchor the legend to the top left
    )
)
# Optionally, if you want to remove axis ticks and labels you can do so by setting ticks to '' and showticklabels to False
fig.update_xaxes(ticks='outside', showticklabels=True, linewidth=1, linecolor='black')
fig.update_yaxes(ticks='outside', showticklabels=True, linewidth=1, linecolor='black')
fig.update_layout(
    paper_bgcolor='rgba(255,255,255,255)',
    plot_bgcolor='rgba(255,255,255,255)',
    title='Portfolio Strategy Overview'
)
# Display the plot
fig.show()
```
