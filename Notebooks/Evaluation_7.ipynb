## MPT - GARCH

```python
def RMSE(actual, predicted):
    """
    Calculate the Root Mean Square Error (RMSE) between actual and predicted values.
    
    Parameters:
    - actual (array-like): The true values.
    - predicted (array-like): The predicted values.
    
    Returns:
    - Root Mean Square Error (RMSE) value.
    """
    return round(np.sqrt(np.mean((actual - predicted)**2)),5)*100
```

```python
def RMSPE(actual, predicted):
    """
    Calculate the Root Mean Square Percentage Error (RMSPE) between actual and predicted values.
    
    Parameters:
    - actual (array-like): The true values.
    - predicted (array-like): The predicted values.
    
    Returns:
    - Root Mean Square Percentage Error (RMSPE) value.
    """
    rmspe = np.sqrt(np.mean(((actual - predicted) / actual) ** 2))
    return round(rmspe * 100, 5)  # Multiplying by 100 to get the percentage and rounding to 5 decimal places
```

```python
import matplotlib.pyplot as plt

# Define the function to create a 2x3 grid of subplots for each asset's RMSE
def plot_rmse_grid(assets, results_dict, row_names, line_styles, markers):
    # Create a 2x3 grid of subplots
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()  # Flatten the grid to iterate easily

    # Loop through each asset and plot the RMSE in a subplot
    for i, asset in enumerate(assets):
        ax = axes[i]  # Get the current axis for the subplot

        # Concatenate the rows for each asset from different models
        combined_rows = pd.concat([results.iloc[i:i+1, :] for results in results_dict.values()], ignore_index=True)
        combined_rows = combined_rows.iloc[:, 1:-1]  # Exclude the first and last column if necessary

        # Plot each model's RMSE on the subplot
        for index, row in combined_rows.iterrows():
            ax.plot(row.index, row.values, marker=markers[index], linestyle=line_styles[index], label=row_names[index])

        ax.set_title(f'RMSE for {asset}')
        ax.set_xlabel('Year')
        ax.set_ylabel('RMSE')
        ax.legend()

    # Hide any unused subplots if the number of assets is less than 6
    for j in range(i+1, 6):
        fig.delaxes(axes[j])

    plt.tight_layout()  # Adjust the layout
    plt.show()

# Define your line styles, row names, and markers as per your models
line_styles = ['-', '--', '-.', ':', (0, (3, 5, 1, 5)), (0, (3, 1, 1, 1))]
row_names = ['MP-MTL-GARCH', 'TARCH', 'GARCH', 'AVGGARCH', 'EGARCH', 'GJR-GARCH']
markers = ['o', 's', '^', 'D', 'x', '*']

# Dictionary to hold your different models' DataFrames
results_dict = {
    'MPT': results_MPT,
    'Tarch': results_Tarch,
    'Garch': results_garch,
    'Avg': results_avg,
    'E': results_e,
    'GJR': results_GJR
}

# List of assets to plot
assets = ['DHR', 'GILD', 'ISRG', 'JNJ', 'LLY', 'PFE']  # Replace with your actual asset names

# Call the function with your assets and other parameters
plot_rmse_grid(assets, results_dict, row_names, line_styles, markers)
```
